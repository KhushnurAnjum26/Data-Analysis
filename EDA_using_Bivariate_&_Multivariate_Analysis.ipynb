There canbe 3 types of combination for columns-

1-numerical-numerical

2-numerical-categorical

3-categorical-categorical

# **1)Scatterplot(Numerical-Numerical)**

```python
#For bivariate only two argument is passed
#sns.scatterplot(x=tips['total_bill'], y=tips['tip'])

#For multivariate more than two argument is passed
sns.scatterplot(x=tips['total_bill'], y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

# **2)Barplot(Numerical - Categorical)**
it is used when two types o data are given-normally-numerical and categorical

```python
#sns.barplot(x=titanic['pclass'],y=titanic['age'])

sns.barplot(x=titanic['pclass'],y=titanic['age'],hue=titanic['sex'])
```

# **2)Boxplot(Numerical - Categorical)**
it is used when two types o data are given-normally-numerical and categorical


```python
#sns.boxplot(x=titanic['sex'],y=titanic['age'])

sns.boxplot(x=titanic['sex'],y=titanic['age'],hue=titanic['pclass'])
```

# **2)Distplot(Numerical - Categorical)**
it is used when two types o data are given-normally-numerical and categorical

```python
sns.distplot(titanic[titanic['survived']==0]['age'],hist=False)

sns.distplot(titanic[titanic['survived']==1]['age'],hist=False)
```

# **2)Heatmap(Categorical - Categorical)**
it is used when for only categorical types data

```python
import pandas as pd
sns.heatmap(pd.crosstab(titanic['pclass'],titanic['survived']))
```

```python
(titanic.groupby('pclass').mean()['survived'])
```
